```python
import pandas as pd
```

```python
df = pd.read_csv("D:/Pre/others/titanic.csv")
df
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
# Separate out target variable and inputs:

target = df.Survived
inputs = df.drop('Survived', axis='columns')
```

```python
# Dummies to encode male/female into 2 different columns as 0/1 :

dummies = pd.get_dummies(inputs.Sex)
dummies.head(10)
```

```python
# Concat dummies with inputs:

inputs = pd.concat([inputs,dummies],axis='columns')
inputs
```

```python
# Drop sex column:

inputs.drop('Sex',axis='columns',inplace=True)
inputs.head(5)
```

```python
# Find is there any column without number value:

inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
# If not a number then fill it with the mean of all Ages:

inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head(10)
```

```python
# Split dataset into training and testing:

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, train_size = 0.8)
```

```python
len(X_train)
```

```python
# Import Naive Bayes classifier:

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
# Train the model:

model.fit(X_train, y_train)
```

```python
# Find Accuracy:

model.score(X_test, y_test)
```
